# 3 answer
1. subplot() function adds subplot to a current figure at the specified grid position.
2. It is similar to the subplots() function however unlike subplots() it adds one subplot at a time.
3. So to create multiple plots you will need several lines of code with the subplot() function.

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
ax1=plt.subplot(1,1,1)
ax1.plot(x2,y2)
ax1.plot(x2,y2)
ax1.plot(x3,y3)
ax1.plot(x4,y4)
plt.show()
```

```python
fig, ax=plt.subplots()
ax.plot(x1,y1)
ax.plot(x2,y2)
ax.plot(x3,y3)
ax.plot(x4,y4)
plt.show()
```

# 4 answer
Bar plots are a type of data visualization used to represent data in the form of rectangular bars. The height of each bar represents the value of a data point, and the width of each bar represents the category of the data.

A bar chart is used when you want to show a distribution of data points or perform a comparison of metric values across different subgroups of your data. From a bar chart, we can see which groups are highest or most common, and how other groups compare against the others.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,3))
plt.bar(company,profit,color='#0fdfbe')
plt.xlabel("company")
plt.ylabel("profit")
plt.title("company vs profit")
```

```python
# 5 answer
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum.

Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups. They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outliers.

```python
data =[box1,box2]
plt.boxplot(data)
plt.show()
```
